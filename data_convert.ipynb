```python
import pandas as pd
import json

with open('./first_data.json') as json_file:
    data = json.load(json_file)

for item in data:
    entities = item['entities']
    hashtags = item['entities']['hashtags']
    user_mentions = item['entities']['user_mentions']
    urls = item['entities']['urls']
    hashtag_list = list()
    try:
        for hashtag in hashtags:
            del hashtag['indices']
            hashtag_list.append(hashtag['text'])
    except:
        pass
    entities.update({'hashtags':hashtag_list})
    user_mention_screen_name_list = list()
    user_mention_name_list = list()
    user_mention_id_list = list()
    try:
        for user_mention in user_mentions:
            del user_mention['id_str']
            del user_mention['indices']
            user_mention_screen_name_list.append(user_mention['screen_name'])
            user_mention_name_list.append(user_mention['name'])
            user_mention_id_list.append(user_mention['id'])
    except:
        pass              
    entities.update({'user_mentions':{
        'screen_names':user_mention_screen_name_list,
        'names':user_mention_name_list,
        'id':user_mention_id_list
    }})
    url_list = list()
    expanded_url_list = list()
    display_url_list = list()
    try:
        for url in urls:
            del url['indices']
            url_list.append(url['url'])
            expanded_url_list.append(url['expanded_url'])
            display_url_list.append(url['display_url'])
    except:
        pass
    entities.update({'urls':{
        'url':url_list,
        'expanded_url':expanded_url_list,
        'display_url':display_url_list
    }})

df = pd.json_normalize(data, sep='_', errors='ignore')
df = df.drop(columns=['entities_media', 'extended_entities_media'], axis=1)
pd.set_option('display.max_colwidth', None)
with pd.option_context("display.max_columns", None, "display.max_rows", None):
    display(df)
df.to_parquet('/Users/rishabh/Desktop/practice/test_files/final_flatten_data.parquet')
```
