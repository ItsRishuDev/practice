```python
import pandas as pd
import json
from pandas.io.json import json_normalize
# import psycopg2
# import sqlalchemy
# from sqlalchemy import create_engine

with open('./first_data.json') as json_file:
    data = json.load(json_file)

df = pd.json_normalize(data, sep='_', errors='ignore')

df = df.explode('entities_hashtags').reset_index(drop=True)
df = df.explode('entities_symbols').reset_index(drop=True)
df = df.explode('entities_user_mentions').reset_index(drop=True)
df = df.explode('entities_urls').reset_index(drop=True)
df = df.explode('entities_media').reset_index(drop=True)
cols_to_normalize = ['entities_hashtags', 'entities_symbols', 'entities_user_mentions', 'entities_urls', 'entities_media']

normalized = list()
for col in cols_to_normalize:
    
    d = pd.json_normalize(df[col], sep='_')
    d.columns = [f'{col}_{v}' for v in d.columns]
    normalized.append(d.copy())

df = pd.concat([df] + normalized, axis=1)
# df = df.drop(columns=cols_to_normalize, axis=1)

# df.to_parquet('/Users/rishabh/Desktop/practice/test_files/flatten_data_single_column.parquet')
pd.set_option('display.max_colwidth', None)
with pd.option_context("display.max_columns", None, "display.max_rows", None):
    display(df)
```
