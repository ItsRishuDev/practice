```python
import pandas as pd
import json
import psycopg2
import sqlalchemy
from sqlalchemy import create_engine

with open('./data.json') as json_file:
    data = json.load(json_file)

final_data = []
for each in data:
    filtered_data = each['content']['itemContent']['tweet_results']['result']['legacy']
    final_data.append(filtered_data)


df = pd.json_normalize(final_data)

df

 
```

```python
conn_string = 'postgresql://postgres:1234@localhost/tweets'
db = create_engine(conn_string)
conn = db.connect()
```

```python
df.to_sql('second_data', con=conn, if_exists='replace',
          index=False, dtype={"entities.media": sqlalchemy.types.JSON, "entities.urls": sqlalchemy.types.JSON, "extended_entities.media": sqlalchemy.types.JSON})
```
